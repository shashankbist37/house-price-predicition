# Missing data



```python
total =House_sales.isnull().sum().sort_values(ascending=False)
percent=(House_sales.isnull().sum()/House_sales.isnull().count()).sort_values(ascending=False)

missingdata=pd.concat([total,percent],axis=1,keys=['total','percent'])
missingdata.head(20)
```

```python
#dealing with missing data       #doubt here why missing data before missing data
                                    
House_sales = House_sales.drop((missingdata[missingdata['total'] > 0]).index,1)
House_sales.isnull().sum()
```

```python
House_sales.isnull().sum().max()
```

```python
#standardizing data
saleprice_scaled = StandardScaler().fit_transform(House_sales['SalePrice'][:,np.newaxis]);
low_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][ :10]
high_range= saleprice_scaled[saleprice_scaled[:,0].argsort()][-10:]
print('outer range (low) of the distribution:')
print(low_range)
print('\nouter range (high) of the distribution:')
print(high_range)
```

```python
#bivariate analysis saleprice/grlivarea
var = 'GrLivArea'
data = pd.concat([House_sales['SalePrice'], House_sales[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
```
